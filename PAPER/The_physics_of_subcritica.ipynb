# Analisi Dettagliata: Fisica dei Sistemi Moltiplicanti Sottocritici

Questa analisi si basa sull'articolo di **Augusto Gandini e Massimo Salvatores**, pubblicato sul *Journal of Nuclear Science and Technology* (Vol. 39, No. 6, 2002). Il documento esplora la fisica dei reattori guidati da acceleratore (ADS - Accelerator Driven Systems), focalizzandosi sulla neutronica del nocciolo sottocritico, la cinetica, il controllo e la validazione sperimentale.

---

## I. Introduzione
Gli ADS sono riconosciuti a livello mondiale come strumenti fondamentali per la trasmutazione delle scorie radioattive (Attinidi Minori e Plutonio), riducendo il carico sui depositi geologici.

* **Contesto di Sicurezza:** Le strategie di trasmutazione richiedono noccioli con spettro veloce e combustibili privi di materiale fertile (Uranio-238), dominati da Plutonio e Attinidi Minori.
* **Problematiche dei noccioli critici:** Questi combustibili comportano una frazione di neutroni ritardati ($\beta_{eff}$) molto bassa e un coefficiente di reattività Doppler basso o quasi nullo.
* **Soluzione ADS:** La sottocriticità del sistema permette di mitigare o eliminare le conseguenze negative di queste caratteristiche di sicurezza.

---

## II. Il Nocciolo Sottocritico in Regime Stazionario

### 1. Distribuzione del Flusso
In un sistema critico, il bilancio neutronico è descritto dall'equazione omogenea di Boltzmann. In un sistema sottocritico (ADS), l'equilibrio stazionario è mantenuto da una sorgente esterna $s$. L'equazione che governa il sistema diventa disomogenea:
$$A\phi_{in} + P\phi_{in} + s = 0$$
Dove $\phi_{in}$ è il flusso disomogeneo, $A$ è l'operatore di assorbimento/trasporto e $P$ è l'operatore di produzione.

### 2. Definizioni di Reattività
Il documento discute l'ambiguità nella definizione della moltiplicazione in sistemi guidati da sorgente:
* **$K_{eff}$ (Standard):** L'autovalore che bilancia l'equazione omogenea (senza sorgente).
* **$K_s$ (Source Multiplication):** Definito come il rapporto tra neutroni prodotti da fissione e la somma di neutroni da fissione più quelli di sorgente.
    $$K_s = \frac{\langle u, P\phi_{in} \rangle}{\langle u, P\phi_{in} \rangle + \langle u, s \rangle}$$
* **$K_{sub}$ (Power-weighted):** Una definizione più fisica che usa una funzione importanza ($n^*_o$) legata alla potenza del reattore, tenendo conto che i neutroni di sorgente e di fissione hanno "importanze" diverse.

### 3. Il Parametro $\varphi^*$
Questo è un parametro cruciale per gli ADS. Rappresenta il rapporto tra l'importanza media dei neutroni di sorgente e l'importanza media dei neutroni di fissione.
La relazione fondamentale che lega la corrente del fascio protonico ($i_p$) alla potenza termica ($W$) e alla sottocriticità è:
$$i_p (Ampere) = \frac{\bar{\nu} (1/K_{eff} - 1)}{\varphi^* Z} \cdot \frac{W}{\bar{\epsilon}_f}$$
Dove $Z$ è il numero di neutroni per protone incidente. Un valore di $\varphi^* > 1$ riduce la richiesta di corrente dell'acceleratore.

## III. Cinetica di un Sistema Sottocritico

### 1. Equazioni della Cinetica Puntuale
Vengono derivate le equazioni cinetiche per un sistema sottocritico in termini di potenza normalizzata $P_N$. A differenza dei sistemi critici, le equazioni includono termini espliciti per la sorgente esterna ($P_{source}$) e la "reattività generalizzata" ($\rho_{gen}$).

### 2. Comportamento Asintotico
In un sistema sottocritico, un inserimento di reattività positiva $\Delta \rho$ (purché il sistema resti sottocritico) porta a un nuovo livello di potenza di equilibrio, non a una divergenza esponenziale come nei reattori critici.
* *Esempio:* Se un sistema è sottocritico di $-10\$$ e si inseriscono $+5\$$, la potenza raddoppia, ma si stabilizza.
* *Scala temporale:* La cinetica è dominata dalla variazione della sorgente esterna, che avviene su tempi scala dei microsecondi (vita media dei neutroni pronti), molto più velocemente dei tempi dei neutroni ritardati.

### 3. Incidenti di Reattività e Loss-of-Flow
* **Inserzione di Reattività:** L'aumento di potenza in un ADS è molto più lento e limitato rispetto a un reattore critico, riducendo l'energia totale rilasciata in un transitorio accidentale.
* **Loss-of-Flow (LOF):** Se il flusso del refrigerante si ferma ma la sorgente non viene spenta, un ADS potrebbe comportarsi peggio di un reattore critico, poiché mancano i feedback negativi forti che spegnerebbero un reattore critico "standard". Questo sottolinea l'importanza di spegnere la sorgente (beam trip) in caso di incidente.

### 4. Scelta del Livello di Sottocriticità
Non esiste un criterio unico, ma un compromesso tra:
* **Sicurezza:** Il sistema deve rimanere sottocritico anche passando dallo stato "caldo" (operativo) allo stato "freddo" (spento), considerando anche eventuali inserzioni accidentali di reattività (es. svuotamento del sodio).
    $$K_{eff} + \Delta K_{feedback} + \Delta K_{accident} < 1$$
* **Efficienza:** Una sottocriticità troppo profonda richiede correnti dell'acceleratore troppo elevate (costose) e provoca una distribuzione di potenza molto piccata spazialmente (peaking factor).

### 5. Controllo e Monitoraggio
* **Controllo:** Si discute se usare barre di controllo o variare solo l'intensità del fascio. Variare solo il fascio richiederebbe escursioni di corrente molto ampie (fattore 5) per compensare il burn-up.
* **Monitoraggio:** Tecniche proposte includono l'uso della sorgente in modalità pulsata (analizzando il decadimento dei neutroni) o il metodo MSM (Modified Source Multiplication) calibrato vicino alla criticità.

## IV. Validazione Sperimentale

La fisica degli ADS necessita di validazione sperimentale, in particolare per quanto riguarda il parametro $\varphi^*$ e la transizione tra regime dominato dalla sorgente e regime dominato dai feedback.

### Programma MUSE (MASURCA, Cadarache)
Gli esperimenti MUSE utilizzano il reattore veloce MASURCA accoppiato all'acceleratore di deuteroni GENEPI per produrre neutroni tramite reazioni (d,d) o (d,t).
* **Risultato Chiave:** Simulazioni e misure confermano che lo spettro neutronico all'interno del nocciolo (vicino alla zona buffer) è molto simile sia usando una sorgente di spallazione reale sia usando neutroni da (d,t), validando l'approccio sperimentale.
* **Misure:** Sono state misurate distribuzioni di flusso radiali e assiali, evidenziando il "peaking" dovuto alla sorgente esterna.

## V. Immagini di ADS e Problemi Tecnologici

Il documento presenta diversi design concettuali:
1.  **Energy Amplifier (C. Rubbia):** Raffreddato a Piombo, con target di spallazione integrato.
2.  **Progetto Ansaldo (XADS):** Raffreddato a metallo liquido (LBE o Pb), dimostratore da 80-100 MWt.
3.  **Progetto Framatome:** Raffreddato a Gas.

**Problemi Tecnologici Aperti:**
* **Schermaggio:** I neutroni ad alta energia ($>100$ MeV) da spallazione richiedono schermature massicce nella parte superiore del reattore per penetrazione profonda.
* **Finestra del Fascio:** L'ingresso verticale del fascio complica il sistema di caricamento del combustibile e pone problemi di sicurezza per il rischio di rottura della finestra e ingresso di neutroni di backscattering nel tubo a vuoto.

---

## VI. Conclusioni

La fisica del nocciolo dei sistemi sottocritici è generalmente ben compresa. Tuttavia, la validazione sperimentale di parametri specifici (come $\varphi^*$) e del comportamento dinamico è essenziale. Gli esperimenti in corso (come MUSE) e futuri dimostratori (come il progetto TRIGA ipotizzato o l'XADS europeo) sono necessari per consolidare il design di un ADS robusto e sicuro.
