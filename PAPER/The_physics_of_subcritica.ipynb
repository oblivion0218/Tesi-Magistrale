# Analisi Dettagliata: Fisica dei Sistemi Moltiplicanti Sottocritici

Questa analisi si basa sull'articolo di **Augusto Gandini e Massimo Salvatores**, pubblicato sul *Journal of Nuclear Science and Technology* (Vol. 39, No. 6, 2002). Il documento esplora la fisica dei reattori guidati da acceleratore (ADS - Accelerator Driven Systems), focalizzandosi sulla neutronica del nocciolo sottocritico, la cinetica, il controllo e la validazione sperimentale.

---

## I. Introduzione
Gli ADS sono riconosciuti a livello mondiale come strumenti fondamentali per la trasmutazione delle scorie radioattive (Attinidi Minori e Plutonio), riducendo il carico sui depositi geologici.

* **Contesto di Sicurezza:** Le strategie di trasmutazione richiedono noccioli con spettro veloce e combustibili privi di materiale fertile (Uranio-238), dominati da Plutonio e Attinidi Minori.
* **Problematiche dei noccioli critici:** Questi combustibili comportano una frazione di neutroni ritardati ($\beta_{eff}$) molto bassa e un coefficiente di reattività Doppler basso o quasi nullo.
* **Soluzione ADS:** La sottocriticità del sistema permette di mitigare o eliminare le conseguenze negative di queste caratteristiche di sicurezza.

---

## II. Il Nocciolo Sottocritico in Regime Stazionario

### 1. Distribuzione del Flusso
In un sistema critico, il bilancio neutronico è descritto dall'equazione omogenea di Boltzmann. In un sistema sottocritico (ADS), l'equilibrio stazionario è mantenuto da una sorgente esterna $s$. L'equazione che governa il sistema diventa disomogenea:
$$A\phi_{in} + P\phi_{in} + s = 0$$
Dove $\phi_{in}$ è il flusso disomogeneo, $A$ è l'operatore di assorbimento/trasporto e $P$ è l'operatore di produzione.

### 2. Definizioni di Reattività
Il documento discute l'ambiguità nella definizione della moltiplicazione in sistemi guidati da sorgente:
* **$K_{eff}$ (Standard):** L'autovalore che bilancia l'equazione omogenea (senza sorgente).
* **$K_s$ (Source Multiplication):** Definito come il rapporto tra neutroni prodotti da fissione e la somma di neutroni da fissione più quelli di sorgente.
    $$K_s = \frac{\langle u, P\phi_{in} \rangle}{\langle u, P\phi_{in} \rangle + \langle u, s \rangle}$$
* **$K_{sub}$ (Power-weighted):** Una definizione più fisica che usa una funzione importanza ($n^*_o$) legata alla potenza del reattore, tenendo conto che i neutroni di sorgente e di fissione hanno "importanze" diverse.

### 3. Il Parametro $\varphi^*$
Questo è un parametro cruciale per gli ADS. Rappresenta il rapporto tra l'importanza media dei neutroni di sorgente e l'importanza media dei neutroni di fissione.
La relazione fondamentale che lega la corrente del fascio protonico ($i_p$) alla potenza termica ($W$) e alla sottocriticità è:
$$i_p (Ampere) = \frac{\bar{\nu} (1/K_{eff} - 1)}{\varphi^* Z} \cdot \frac{W}{\bar{\epsilon}_f}$$
Dove $Z$ è il numero di neutroni per protone incidente. Un valore di $\varphi^* > 1$ riduce la richiesta di corrente dell'acceleratore.

## **III. La Cinetica di un Sistema Sottocritico**

### **1. Equazione Base e Cinetica Puntuale**

La cinetica di un **Sistema Moltiplicante Sottocritico (SC)** è governata dalle equazioni che descrivono l'evoluzione temporale del flusso neutronico e delle densità dei precursori di neutroni ritardati, tenendo conto della **sorgente esterna** $s$.

Le equazioni base accoppiate per il vettore del flusso neutronico $\phi$ (multigruppo) e il vettore delle densità dei precursori $c_i$ sono:

$$V^{-1}\frac{d\phi}{dt}=A\phi+(1-\beta)\chi_{P}F\phi+\chi_{D}u\sum_{i=1}^{I}\lambda_{i}c_{i}+s$$
$$\frac{dc_{i}}{dt}=\beta_{i}v\Sigma_{f}^{T}\phi-\lambda_{i}c_{i}$$
$$(\beta=\sum_{i=1}^{I}\beta_{i})$$

Dove:
* $V^{-1}$ è la matrice diagonale inversa delle velocità neutroniche.
* $A$ è l'operatore di perdita, assorbimento e scattering.
* $F$ è l'operatore di produzione da fissione.
* $\chi_P$ e $\chi_D$ sono gli spettri dei neutroni pronti e ritardati.
* $u$ è un vettore unitario.
* $\beta_i$ e $\lambda_i$ sono la frazione e la costante di decadimento dei precursori.

Seguendo la derivazione (basata sull'uso della funzione di importanza generalizzata $n_o^*$), le equazioni possono essere riformulate in termini di **Potenza Normalizzata** $P_N$ (dove $P_N(t=0)=1$):

$$\frac{dP_{N}}{l_{eff}dt} = (\rho_{gen} - \alpha\beta) \frac{P_{N}}{l_{eff}} + \alpha\sum_{i=1}^{I}\lambda_{i}\xi_{i} + \zeta(1-P_{N}) + \rho_{source}$$
$$\frac{d\xi_{i}}{dt} = \beta_{i}\frac{P_{N}}{l_{eff}} - \lambda_{i}\xi_{i}$$

Dove:
* $\xi_i$ è la densità efficace del precursore $i$-esimo normalizzata.
* $l_{eff}$ è la **vita media effettiva** dei neutroni pronti.
* $\rho_{gen}$ è la **reattività generalizzata** (associata a una perturbazione di sistema $\delta A, \delta F$).
* $\rho_{source}$ è la **reattività della sorgente** (associata a una perturbazione della sorgente $\delta s$).
* $\zeta$ è l'**indice di sottocriticità**, definito come $\zeta = \frac{1}{I} \equiv \frac{1-K_{sub}}{K_{sub}}$.
* $\alpha$ è il coefficiente che tiene conto della distribuzione dei neutroni ritardati.

Queste sono le equazioni della **cinetica puntuale** per un sistema sottocritico. Al limite, se il sistema si avvicina alla criticità e la sorgente esterna tende a zero, queste equazioni si riducono alle classiche equazioni della cinetica puntuale per un reattore critico.

---

### **2. Comportamento Asintotico e Scala Temporale**

#### **Potenza Asintotica in Stato Stazionario**

Il comportamento asintotico ($t \rightarrow \infty$) della potenza, a seguito di una perturbazione che mantiene il sistema sottocritico, si ottiene ponendo le derivate temporali a zero ($\frac{dP_N}{dt}=0$, $\frac{d\xi_i}{dt}=0$).

Dalla seconda equazione si ottiene $\xi_i = \beta_{i,eff}/\lambda_i$ (se $\beta_{i,eff} \equiv \beta_i$, ma in realtà le $\beta_i$ sono $\beta_{i,eff}$). Sostituendo nella prima, si ha:

$$\rho_{gen}P_{N}+\zeta(1-P_{N})+\rho_{source}=0$$

da cui si ricava la potenza normalizzata asintotica:

$$P_{N}=\frac{\zeta + \rho_{source}}{\zeta-\rho_{gen}} = \frac{1-K_{sub}+K_{sub}\rho_{source}}{1-K_{sub}\rho_{gen}}$$

* La potenza normalizzata $P_N$ **aumenta** all'aumentare di $\rho_{source}$ (sorgente crescente) e all'aumentare di $\rho_{gen}$ (reattività positiva).
* Se il sistema **si avvicina alla criticità** ($\rho_{gen} \rightarrow \zeta = (1-K_{sub})/K_{sub}$), la potenza **diverge**.

#### **Differenze con Sistemi Critici**

In un sistema sottocritico:
* Un'inserzione di reattività positiva che non raggiunge la criticità (es. un inserimento di $+5\$$ in un sistema a $-10\$$) produce un **raddoppio della potenza**, ma essa si stabilizza, a differenza di un reattore critico che divergerebbe (diventerebbe **prompt critical**).
* Il comportamento cinetico è caratterizzato dai **tempi scala della sorgente esterna** (ordine della **vita media dei neutroni pronti**, tipicamente microsecondi), non dai tempi dei neutroni ritardati (ordine dei $10$ secondi). Una variazione istantanea della sorgente ha un effetto quasi istantaneo sulla potenza.

---

### **3. Sicurezza e Coefficienti di Reattività**

#### **Effetti di Feedback e Livello di Sottocriticità**

L'evoluzione della potenza e della temperatura induce **feedback di reattività** (effetto Doppler, espansione del combustibile, ecc.).
* In un **SC fortemente sottocritico** (regime **"source-dominated"**), il comportamento dinamico è dominato dalla sorgente esterna e le variazioni di reattività da feedback hanno una rilevanza inferiore.
* Quando il sistema è **più vicino alla criticità** (regime **"feedback-dominated"**), gli effetti di feedback diventano più importanti, e il comportamento si avvicina a quello di un reattore critico.

#### **Incidenti di Inserzione di Reattività**

Gli inserimenti rapidi di reattività positiva hanno conseguenze diverse:
* In un SC, l'aumento di potenza è **considerevolmente più lento** e l'energia totale rilasciata è **molto più piccola** rispetto al corrispondente reattore critico.
* *Esempio:* In un transitorio di reattività rapida (rampa di $170\$$/s), un reattore critico raggiunge la criticità pronta e picchi di potenza molto elevati (fino a $700$ GW). Lo stesso nucleo sottocritico di $-3\$$ produce picchi di potenza molto inferiori ($2.2$ GW). 

#### **Incidenti di Perdita di Flusso (Loss-of-Flow, LOF)**

* Se si verifica un incidente di LOF **senza spegnimento della sorgente**, un sistema sottocritico (es. $-10\beta$) può essere **meno favorevole** di un sistema critico.
* Questo perché il reattore critico ha feedback di reattività più efficaci che ne rallentano e riducono l'aumento di temperatura.
* Ciò evidenzia che lo **spegnimento della sorgente (beam trip)** deve essere un elemento di sicurezza essenziale per gli ADS.

#### **Effetto Doppler Ridotto**

Nei sistemi ADS dedicati alla trasmutazione, il combustibile è dominato dagli attinidi minori (MA) in assenza di ${}^{238}\text{U}$, risultando in un **effetto Doppler ridotto**.
* A **sottocriticità profonda**, la differenza tra un coefficiente Doppler "standard" e uno "basso" non è significativa per l'evoluzione della potenza.
* **Vicino alla criticità**, l'effetto Doppler ridotto diventa invece **significativo**. 

---

### **4. Scelta del Livello di Sottocriticità**

Non esiste un criterio finale per un livello di sottocriticità "ottimale". È necessario trovare un **compromesso** tra il regime "source-dominated" (sicurezza) e il regime "feedback-dominated" (efficienza).

#### **Criteri di Sicurezza (Requisiti sul $K_{eff}$)**

Se non sono previste barre di controllo, il livello di sottocriticità deve garantire che il nucleo rimanga sottocritico in tutte le condizioni operative e accidentali:

1.  **Da caldo a freddo (hot $\rightarrow$ cold) + Accidente:**
    Il nucleo freddo deve rimanere sottocritico anche con un'inserzione accidentale di reattività (es. svuotamento del refrigerante $\Delta K_{AC}$). Poiché i feedback termici ($\Delta K_{FB}$) inducono generalmente una reattività positiva passando da caldo a freddo, si richiede:

    $$K_{eff} + \Delta K_{FB} + \Delta K_{AC} < 1$$

2.  **Durante l'Operazione (massima reattività accidentale):**
    Il $K_{eff}$ operativo deve tenere conto della massima reattività accidentale prevedibile ($\Delta K_{AC}^M$):

    $$K_{eff} + \Delta K_{AC}^{M} < 1$$

3.  **Durante l'Operazione con Burn-up:**
    La variazione di reattività dovuta al burn-up ($\Delta K_{BU}$) è generalmente negativa, ma può essere positiva (specialmente con combustibile di attinidi minori che agiscono da "fertilizzanti", come ${}^{241}\text{Am}$). In quel caso, si richiede:

    $$K_{eff} + \Delta K_{AC}^{M} + \Delta K_{BU} < 1$$

#### **Criteri di Efficienza (Compromesso)**

Una sottocriticità **troppo profonda** non è ottimale per ragioni economiche e di performance:
* **Costo della Sorgente:** Forte sorgente esterna implica costi elevati e riduce l'efficienza dell'impianto.
* **Distribuzione di Potenza:** Un nucleo profondamente sottocritico ha una distribuzione di potenza spazialmente **piccata** (dominata dalla sorgente), non ideale per ottimizzare l'irraggiamento e la trasmutazione del combustibile (che richiede un profilo piatto).

---

### **5. Controllo e Monitoraggio della Reattività**

#### **Controllo della Reattività**

* **Solo Sorgente Esterna:** In linea di principio, il controllo di reattività (es. compensazione del burn-up) può essere fatto variando solo l'intensità del fascio protonico ($i_p$).
* **Fattore di Variazione:** Per compensare i cambiamenti di $K_{eff}$ tra lo stato "freddo" ($K_{eff}=0.99$) e la fine del ciclo "caldo" ($K_{eff}=0.95$), la corrente della sorgente dovrebbe variare di un fattore $\approx 5$. Questa ampia escursione rende l'uso di **barre di controllo** una soluzione da esaminare attentamente per alcune funzioni.
* **Overshoot:** L'effetto inverso dello spegnimento del fascio (rapida riduzione di potenza) è un **overshoot** di potenza (aumento istantaneo) dovuto a un aumento improvviso della sorgente. Sebbene più limitato di un'estrazione incontrollata di barre in un reattore critico, deve essere analizzato.

#### **Monitoraggio del Livello di Sottocriticità**

Diversi metodi possono essere utilizzati per misurare e monitorare il livello di sottocriticità $\rho = (1-K_{eff})/K_{eff}$:

1.  **Metodo ad Impulsi (Pulse Mode):**
    * La sorgente viene pulsata.
    * La cinetica puntuale prevede che il decadimento rapido della popolazione neutronica dopo l'impulso sia di tipo $exp(-\alpha t)$, dove $\alpha = (\rho - \beta)/l$.
    * Misurando la costante di decadimento $\alpha$ e conoscendo $\beta$ e $l$, si può estrarre la reattività $\rho$. 

2.  **Moltiplicazione Modificata della Sorgente (Modified Source Multiplication, MSM):**
    * Un reattore sottocritico moltiplica i neutroni della sorgente secondo $S/(1-K_{eff})$.
    * Il conteggio $C$ di un rivelatore è dato da $C = \frac{\epsilon S \delta}{1-K_{eff}}$, dove $\epsilon$ è l'efficienza del rivelatore.
    * Il metodo MSM permette di confrontare il rapporto dei conteggi $C_1/C_2$ tra due stati e correggere per le variazioni di efficienza e sorgente. Se uno stato $K_{eff,1}$ è misurato con un metodo assoluto (es. rod-drop), si possono calcolare tutti gli altri.

3.  **Variazione Sincronizzata di Barra e Sorgente:**
    * Si osserva una variazione di inserzione di barra di controllo $\delta K_{eff} / K_{eff}$ e una variazione di intensità di sorgente $\delta S_{exp}/S$ tale da **mantenere inalterato il livello di potenza**.
    * In queste condizioni, non troppo lontano dalla criticità, si ottiene la relazione:

    $$\frac{\delta K_{eff}}{K_{eff}} + \frac{1}{1-K_{eff}} \frac{\delta S_{exp}}{S} \approx 0$$

    * Questa relazione consente una facile valutazione della sottocriticità $(1-K_{eff})$.

## IV. Validazione Sperimentale

La fisica degli ADS necessita di validazione sperimentale, in particolare per quanto riguarda il parametro $\varphi^*$ e la transizione tra regime dominato dalla sorgente e regime dominato dai feedback.

### Programma MUSE (MASURCA, Cadarache)
Gli esperimenti MUSE utilizzano il reattore veloce MASURCA accoppiato all'acceleratore di deuteroni GENEPI per produrre neutroni tramite reazioni (d,d) o (d,t).
* **Risultato Chiave:** Simulazioni e misure confermano che lo spettro neutronico all'interno del nocciolo (vicino alla zona buffer) è molto simile sia usando una sorgente di spallazione reale sia usando neutroni da (d,t), validando l'approccio sperimentale.
* **Misure:** Sono state misurate distribuzioni di flusso radiali e assiali, evidenziando il "peaking" dovuto alla sorgente esterna.

## V. Immagini di ADS e Problemi Tecnologici

Il documento presenta diversi design concettuali:
1.  **Energy Amplifier (C. Rubbia):** Raffreddato a Piombo, con target di spallazione integrato.
2.  **Progetto Ansaldo (XADS):** Raffreddato a metallo liquido (LBE o Pb), dimostratore da 80-100 MWt.
3.  **Progetto Framatome:** Raffreddato a Gas.

**Problemi Tecnologici Aperti:**
* **Schermaggio:** I neutroni ad alta energia ($>100$ MeV) da spallazione richiedono schermature massicce nella parte superiore del reattore per penetrazione profonda.
* **Finestra del Fascio:** L'ingresso verticale del fascio complica il sistema di caricamento del combustibile e pone problemi di sicurezza per il rischio di rottura della finestra e ingresso di neutroni di backscattering nel tubo a vuoto.

---

## VI. Conclusioni

La fisica del nocciolo dei sistemi sottocritici è generalmente ben compresa. Tuttavia, la validazione sperimentale di parametri specifici (come $\varphi^*$) e del comportamento dinamico è essenziale. Gli esperimenti in corso (come MUSE) e futuri dimostratori (come il progetto TRIGA ipotizzato o l'XADS europeo) sono necessari per consolidare il design di un ADS robusto e sicuro.
