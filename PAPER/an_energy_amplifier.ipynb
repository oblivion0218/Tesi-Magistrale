# 1. Introduzione
- Introduci il concetto di **Energy Amplifier (EA)** usando l’analogia con i calorimetri nella fisica delle alte energie.  
- Presenta l’EA come un **sistema sottocritico** alimentato da un **acceleratore di protoni**.  
- Confrontalo con:  
  - **Reattori convenzionali** (problemi di sicurezza, rischio di meltdown, scorie a lunga vita).  
  - **Fusione** (ancora lontana da un’applicazione industriale).  
- Introduci il **Torio** come combustibile preferibile: abbondante, distribuito globalmente, e con minore produzione di attinidi transuranici.

# 2. Il Torio come combustibile fertile
- Spiega la fisica nucleare del ciclo **${}^{232}Th \rightarrow {}^{233}U$**.  
- Presenta le **equazioni differenziali** che descrivono:  
  - La produzione di ${}^{233}U$ tramite cattura neutronica.  
  - Il decadimento di ${}^{233}Pa$ in ${}^{233}U$.  
- Mostra che, in condizioni stazionarie, si raggiunge un **equilibrio stabile** tra isotopi fertili e fissili sotto flusso neutronico.  
- Nota chiave: discuti perché un **reattore critico al Torio** è difficile da realizzare (margine neutronico troppo stretto), motivando la necessità di una **sorgente esterna di neutroni**.

# 3. La sorgente esterna di neutroni
- Descrivi la **moltiplicazione neutronica assistita da sorgente esterna** in un nocciolo sottocritico.  
- Definisci il **guadagno energetico** \( G \).  
- Analizza l’economia neutronica necessaria a mantenere il sistema sottocritico ma con **alto guadagno**.  
- Valuta l’impatto dei **veleni da fissione** sull’inventario neutronico e sull’operatività.

# 4. Il fascio ad alta energia
- Dettaglia l’interazione tra il **fascio di protoni** e il **bersaglio (target)**.  
- Spiega il processo di **spallazione** responsabile della produzione dei neutroni primari.  
- Confronta i moderatori:  
  - **Acqua** (in configurazione sotto-moderata per limitare catture parassite).  
  - **Grafite** (moderazione più “pulita” e migliore gestione termica).  
- Affronta il problema del contenimento della **cascata ad alta energia** e dei vincoli sulla geometria del target.

# 5. L’acceleratore di particelle ad alta energia
- Presenta le specifiche tecniche necessarie per l’acceleratore.  
- Confronta due soluzioni principali:  
  - **LINAC** (acceleratore lineare)  
  - **Ciclotrone Isocrono**  
- Definisci i requisiti operativi:  
  - Corrente di fascio ~ **7 mA**  
  - Energia del fascio **800 MeV – 1 GeV**  
- Collega questi parametri alla **potenza termica** ottenibile nel nocciolo.

# 6. Il carico iniziale di combustibile
- Discute le strategie di avviamento (“**priming the pump**”).  
- Poiché il torio naturale non è fissile, serve un carico iniziale di:  
  - ${}^{233}U$, oppure  
  - ${}^{235}U$, oppure  
  - ${}^{239}Pu$  
- Descrivi come raggiungere l’**equilibrio secolare** del ciclo torico sotto irraggiamento.

# 7. Alcune possibilità pratiche
- Presenta due design ingegneristici di riferimento:  
  - **EA raffreddato ad acqua**  
    - Simile a un PWR  
    - Nessun target separato (il fascio colpisce direttamente moderatore/combustibile)  
  - **EA raffreddato a gas (Elio)**  
    - **Target di Pb-Bi** separato  
    - Moderazione a **grafite**  
    - Configurazione simile ai reattori *pebble-bed*.

# 8. Considerazioni commerciali
- Analizza gli aspetti economici preliminari: costo dell’acceleratore, complessità operativa, scalabilità.  
- Impatto ambientale: forte riduzione delle **scorie nucleari a lunga vita**.

# 9. Sintesi
- Ricapitola i punti chiave:  
  - L’EA è un sistema **intrinsecamente sicuro** grazie alla sottocriticità.  
  - Il ciclo al Torio sostenuto da una sorgente esterna risolve il deficit neutronico dei reattori critici.  
  - Le sfide tecnologiche sono importanti, ma il potenziale energetico a lungo termine è significativo.
