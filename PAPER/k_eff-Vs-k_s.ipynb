### **Introduzione (Capitolo 1)**

Questo capitolo definisce il contesto e gli obiettivi dello studio, introducendo i parametri fondamentali per l'analisi dei sistemi sottocritici guidati da acceleratore (ADS).

* **Contesto dell'ADS:** I sistemi ADS (Accelerator-Driven Systems) sono sviluppati per diverse applicazioni, tra cui la **produzione di energia**, la **trasmutazione delle scorie nucleari** (combustibile esaurito) e l'uso come **sorgente di neutroni**. 
* **Sicurezza e Sottocriticità:** Un ADS è progettato per operare in un livello di **sottocriticità sufficientemente profondo** per garantire la sicurezza intrinseca.
* **Limiti di $k_{eff}$:** Tradizionalmente, la sottocriticità è espressa dal **fattore di moltiplicazione effettivo $k_{eff}$**. Tuttavia, in un sistema ADS, $k_{eff}$ (un autovalore calcolato , Nel contesto della fisica dei reattori, "autovalore calcolato" si riferisce al fatto che $k_{eff}$ è la soluzione (specificamente, l'autovalore fondamentale) di un problema matematico omogeneo, formulato per un sistema privo di sorgenti esterne.Si risolve l'equazione stazionaria del trasporto neutronico (spesso nella sua approssimazione diffusiva) cercando le condizioni per una soluzione stazionaria ($\phi$) senza un termine di sorgente ($S=0$).) non è sufficiente perché **non tiene conto della posizione e dell'energia della sorgente esterna**.
* **Introduzione di $k_s$ e $\phi^*$:** Per superare i limiti di $k_{eff}$ nell'analisi delle prestazioni di un ADS, vengono introdotti due parametri più fisici:
    1.  **Fattore di moltiplicazione subcritico ($k_s$):** Questo parametro è definito come il **rapporto tra i neutroni prodotti da fissione e i neutroni totali** (quelli da fissione più quelli della sorgente esterna) nel sistema. È direttamente correlato alla moltiplicazione neutronica $M$ tramite la nota formula $M = 1 / (1 - k_s)$.
    2.  **Efficienza della sorgente ($\phi^*$):** Questo parametro viene introdotto per correlare $k_s$ al $k_{eff}$ convenzionale. $\phi^*$ agisce come un **indice dell'importanza della sorgente esterna** rispetto alla fissione e quantifica quanto efficacemente i neutroni della sorgente contribuiscono agli eventi di fissione finali e alla potenza totale del sistema.
* **Importanza dei Parametri:** $k_s$ e $\phi^*$ sono fondamentali per la valutazione delle prestazioni dell'ADS. La potenza totale del reattore $P$ può essere espressa in funzione di $k_s$ (Equazione 2). Utilizzando la relazione tra $k_s$ e $k_{eff}$, la potenza può anche essere espressa in funzione di $\phi^*$ (Equazione 3). Quest'ultima equazione mostra che **ottimizzare la sorgente per ottenere un $\phi^*$ più elevato** permette di ridurre la potenza richiesta all'acceleratore per una data potenza del reattore, aumentando l'efficienza del sistema.
* **Obiettivo dello Studio:** Lo studio mira a discutere la relazione teorica tra $k_s$ e $k_{eff}$, investigare numericamente la loro dipendenza dalla sottocriticità e dalle proprietà della sorgente, e infine **esaminare una metodologia sperimentale per misurare $k_s$ e $\phi^*$** utilizzando le distribuzioni dei tassi di reazione misurate negli esperimenti ADS al KUCA.

### **Fattore di Moltiplicazione Subcritico ed Efficienza della Sorgente (Capitolo 2)**

Questo capitolo stabilisce il fondamento teorico, derivando le definizioni formali di $k_s$ e $\phi^*$ e la loro relazione.

#### **2.1. Fattore di moltiplicazione subcritico $k_s$**

* **Definizione di $k_{eff}$:** Si parte dall'equazione del bilancio neutronico nel modo fondamentale (sistema critico, Eq. 4). Da questa, $k_{eff}$ è definito come il rapporto tra neutroni prodotti (operatore $P$) e neutroni persi (operatore $L$) (Eq. 5).
* **Equazione del Sistema Subcritico:** Per un sistema subcritico *con una sorgente esterna* $s(r, E)$ in stato stazionario, l'equazione di bilancio diventa (Eq. 6):
    $$L\phi_s(r,E) = P\phi_s(r,E) + s(r,E)$$
    dove $\phi_s(r,E)$ è il **flusso neutronico reale** nel sistema con sorgente.
* **Definizione di $k_s$:** Applicando lo stesso concetto di $k_{eff}$ (produzione/perdita) al sistema reale descritto dall'Eq. 6, si definisce $k_s$ (Eq. 7). Integrando l'Eq. 6 e sostituendo, si ottiene la definizione operativa di $k_s$ (Eq. 8):
    $$k_s = \frac{\langle P\phi_s(r,E) \rangle}{\langle L\phi_s(r,E) \rangle} = \frac{\langle P\phi_s(r,E) \rangle}{\langle P\phi_s(r,E) \rangle + \langle s(r,E) \rangle} = \frac{F}{F+S}$$
    dove $F$ è il rateo totale di neutroni da fissione (Eq. 9) e $S$ è il rateo totale di neutroni dalla sorgente (Eq. 10).
* **Concetto Chiave:** $k_s$ è l'analogo di $k_{eff}$ per un sistema sottocritico con sorgente; entrambi descrivono i neutroni di fissione prodotti per neutrone perso nel rispettivo sistema.

#### **2.2. Relazione tra $k_s$ e $k_{eff}$**

* **Differenza tra $k_s$ e $k_{eff}$:**
    * Se il sistema è **molto vicino alla criticità**, il flusso reale $\phi_s$ è simile al flusso fondamentale $\phi_0$, e quindi $k_s \approx k_{eff}$.
    * Se il sistema è **profondamente sottocritico** (come un ADS), $\phi_s$ (determinato dalla sorgente) è molto diverso da $\phi_0$ (il modo fondamentale).
* **Significato Fisico:** $k_{eff}$ (l'autovalore) è calcolato usando un flusso "immaginario" $\phi_0$ che non esiste nel reattore reale, mentre $k_s$ è il parametro *fisico* perché è calcolato usando il flusso *reale* $\phi_s$.
* **Derivazione della Relazione:** Per correlare i due, si introduce il **flusso aggiunto $\phi_0^\dagger$** (noto come "importanza neutronica" del modo fondamentale).
    1.  Si manipola l'equazione aggiunta (Eq. 11) per ottenere $1/k_{eff}$ in termini di flusso reale $\phi_s$ e flusso aggiunto $\phi_0^\dagger$ (Eq. 12).
    2.  Si moltiplica l'Eq. 6 (bilancio con sorgente) per $\phi_0^\dagger$ e si integra (Eq. 13).
    3.  Combinando i passaggi, si arriva all'Eq. 14.
* **Definizione di $\phi^*$:** A questo punto, si definisce formalmente l'efficienza della sorgente $\phi^*$ (Eq. 15):
    $$\phi^* \equiv \frac{\langle \phi_0^\dagger, s \rangle / \langle s \rangle}{\langle \phi_0^\dagger, P\phi_s \rangle / \langle P\phi_s \rangle}$$
    $\phi^*$ è il **rapporto tra l'importanza media della sorgente** (quanto i neutroni della sorgente contribuiscono al processo di moltiplicazione, pesato dal flusso aggiunto) e l'**importanza media della fissione** (quanto i neutroni di fissione contribuiscono).
* **Relazione Finale:** Sostituendo le definizioni (Eq. 8, 12, 16) si ottiene la relazione teorica chiave (Eq. 17):
    $$(1 - \frac{1}{k_{eff}}) = \phi^* (1 - \frac{1}{k_s})$$
    Questa equazione lega il $k_{eff}$ (autovalore del sistema) al $k_s$ (parametro fisico) tramite l'efficienza $\phi^*$ (che dipende dalla geometria e dall'energia della sorgente).

#### **2.3. Misura di $k_s$ e $\phi^*$ mediante distribuzione dei tassi di reazione**

* **Metodologia Sperimentale:** Questo paragrafo propone un metodo per *misurare* $k_s$ e $\phi^*$ sperimentalmente, invece di affidarsi solo ai calcoli.
* **Approssimazione di F e S:** L'idea è di stimare i termini $F$ (fissione totale) e $S$ (sorgente totale) misurando i tassi di reazione di specifici materiali rivelatori (D1 e D2) nelle rispettive regioni (combustibile $V_F$ e targhetta $V_S$) (Eq. 18, 19).
* **Scelta dei Rivelatori:**
    1.  **Per F (D1):** Si usa la reazione **${}^6Li(n,t)\alpha$**. Si assume che la sua sezione d'urto sia ragionevolmente proporzionale a quella di fissione (es. ${}^{235}U(n,f)$) nella regione del combustibile.
    2.  **Per S (D2):** Si usa la reazione **${}^{232}Th(n,f)$**. Questa è una reazione a soglia, adatta per misurare i neutroni veloci (es. 14 MeV da una sorgente D-T) generati nella regione della targhetta.
* **Calcolo Sperimentale di $k_s$:** Sostituendo queste misure (Eq. 18, 19) nella definizione di $k_s$ (Eq. 8), e assumendo la separabilità delle variabili per integrare i tassi di reazione misurati, si ottiene l'Equazione (20). L'Eq. 20 permette di calcolare $k_s$ usando i tassi di reazione misurati ($R_{Li}$, $R_{Th}$) e dei coefficienti di proporzionalità ($C_{Li}$, $C_{Th}$) ottenuti da calcoli MCNPX.
* **Calcolo Sperimentale di $\phi^*$:** $\phi^*$ non viene misurato direttamente. Si ottiene sperimentalmente in modo indiretto:
    1.  Si misura $k_{eff}$ (es. con il metodo *pulsed neutron*).
    2.  Si misura $k_s$ (con il metodo dei tassi di reazione, Eq. 20).
    3.  Si ricava $\phi^*$ usando la relazione teorica (Eq. 17).
