### **An Energy Amplifier ... RUBBIA  1993**

I valori di $k_\infty$ riportati sono il risultato di simulazioni numeriche basate su "computational methods" che integrano le sezioni d'urto nucleari (disponibili all'epoca).

Il calcolo si basa sull'ottimizzazione di due parametri geometrici fondamentali per massimizzare la reattività:
1.  **Raggio dell'elemento di combustibile ($r$):** Dimensione caratteristica delle sfere o barre.
2.  **Rapporto volumetrico ($p$):** Rapporto tra il volume del moderatore e quello del combustibile ($p = V_m / V_f$).

**Condizioni fisiche della simulazione:**
* **Composizione del combustibile:** Miscela all'equilibrio di breeding ${}^{232}Th - {}^{233}U$. La concentrazione di ${}^{233}U$ è quella di regime stazionario, dove il tasso di fissione bilancia il breeding.
* **Definizione di $k_\infty$:** Il fattore è calcolato come $k_\infty = k_{eff} / P$, dove $P$ è la probabilità di non-fuga (ponendo quindi le perdite per leakage a zero).
* **Inclusioni/Esclusioni:** I valori includono le catture parassite nel moderatore. **Non** includono le penalità di reattività ($\Delta k$) dovute a:
    * Accumulo di ${}^{233}Pa$ (dipendente dal flusso).
    * Veleni di fissione saturanti (${}^{135}Xe$, ${}^{149}Sm$) e non saturanti.
    * Isotopi superiori dell'Uranio (${}^{234}U$, ${}^{236}U$).

**Nota sul Berillio:**
Il valore elevato per il Berillio puro ($k_\infty = 1.36$) è teorico e dovuto alla reazione $(n, 2n)$. Tuttavia, il paper nota che l'interazione con i neutroni produce rapidamente ${}^{6}Li$ tramite la reazione $Be(n, \alpha)$. Il ${}^{6}Li$ ha una sezione d'urto di cattura termica enorme ($940$ barn), che satura velocemente portando il $k_\infty$ effettivo a $1.08$.

### **Subcritical multiplication factor ... 2010**

#### 1. Caratteristiche del Reattore (KUCA A-core)
Lo studio è stato condotto presso la **Kyoto University Critical Assembly (KUCA)**, configurata come un sistema guidato da acceleratore (ADS).

* **Combustibile:** Uranio altamente arricchito (HEU - 93% in peso), in lega uranio-alluminio.
* **Moderatore e Riflettore:** Polietilene solido.
* **Sorgente Neutronica:** Generatore di neutroni pulsati da 14 MeV (reazione D-T con bersaglio al trizio), posizionato **esternamente** al nocciolo.
* **Configurazione:** Il nocciolo "A-core" include barre di combustibile standard, barre di controllo (C1-C3) e di sicurezza (S4-S6). Sono stati analizzati 9 casi con diversi gradi di sottocriticità variando il numero di barre di combustibile e l'inserimento delle barre di controllo.


#### 2. Metodologie di Calcolo e Misura
Il paper confronta valori sperimentali (Exp.) e calcolati (Cal.) per $k_{eff}$ e $k_s$:

**A. $k_{eff}$ (Fattore di moltiplicazione effettivo)**
* **Sperimentale:** Misurato con il metodo dei neutroni pulsati (*Pulsed Neutron Method*) utilizzando la **tecnica dell'area estrapolata di Gozani** tramite rivelatori a fibra ottica posizionati nel nocciolo.
    > **Nota sulla Tecnica di Gozani:** Questa tecnica serve a correggere gli effetti spaziali nei sistemi sottocritici. Dopo l'iniezione dell'impulso neutronico, si attende un tempo sufficiente affinché le armoniche spaziali superiori decadano. Successivamente, si fitta la curva di decadimento del modo fondamentale e la si estrapola indietro al tempo zero. Questo permette di calcolare accuratamente il rapporto tra le aree dei neutroni prompt e ritardati, eliminando le distorsioni spaziali iniziali.
* **Calcolo:** Eseguito con il codice Monte Carlo **MCNPX** (libreria ENDF/B-VI.8) in modalità autovalore (*eigenvalue calculation*).

**B. $k_s$ (Fattore di moltiplicazione sottocritico)**
* **Definizione:** $k_s = F / (F+S)$, dove $F$ rappresenta i neutroni da fissione e $S$ i neutroni di sorgente.
* **Sperimentale:** Valutato misurando la distribuzione spaziale dei ratei di reazione tramite rivelatori a fibra ottica mobili (${}^{6}Li$ per la fissione, ${}^{232}Th$ per la sorgente) e applicando un'integrazione numerica.
* **Calcolo:** Eseguito con **MCNPX** in modalità sorgente fissa (*fixed source calculation*).

**Osservazione sui Risultati:**
Si nota una notevole discrepanza tra $k_{eff}$ e $k_s$, con $k_s \ll k_{eff}$. Il paper spiega che questo fenomeno è dovuto alla posizione **esterna** della sorgente neutronica. Molti neutroni di sorgente sfuggono dal sistema senza interagire con il combustibile, riducendo drasticamente l'efficienza di sorgente ($\varphi^* \approx 0.02$) rispetto a una sorgente posizionata al centro del nocciolo.

### **Basic of ADSR   nifenecker**

Nel design degli ADSR esiste un conflitto fondamentale: per avere un alto guadagno energetico ($G$) serve un $k_s$ molto vicino a 1, ma per la sicurezza intrinseca serve un $k_{eff}$ sufficientemente basso (es. $<0.95$).
Il concetto del **Booster** (moltiplicatore a due stadi) risolve questo problema disaccoppiando geometricamente la moltiplicazione della sorgente dall'autovalore del sistema.

**La Fisica del Sistema:**
Il reattore è diviso in due regioni distinte:
1.  **Booster (Stadio 1):** Una piccola sfera centrale di materiale fissile (es. Plutonio puro). Riceve i neutroni della sorgente e li moltiplica.
2.  **Shell (Stadio 2):** Un grande guscio esterno di materiale fertile (es. Torio). Riceve i neutroni in fuga dal Booster e li moltiplica ulteriormente.

Il guadagno totale è il prodotto dei guadagni dei singoli stadi ($M_{tot} \approx M_1 \cdot M_2$), permettendo un'amplificazione enorme pur mantenendo ogni singolo stadio ben al di sotto della criticità.


#### Parametri del Caso Ideale

Di seguito sono riportati i valori teorici per un sistema ottimizzato in cui entrambi gli stadi sono mantenuti a un livello di sicurezza standard ($k \approx 0.95$).

| Parametro | Descrizione | Valore Teorico | Note |
| :--- | :--- | :--- | :--- |
| **$k_1$ (Booster)** | Moltiplicazione della sfera centrale | **0.95** | Stadio sicuro |
| **$k_2$ (Shell)** | Moltiplicazione del guscio esterno | **0.95** | Stadio sicuro |
| **$M_1$** | Guadagno del Booster | **$\approx 20$** | $1/(1-k_1)$ |
| **$M_2$** | Guadagno della Shell | **$\approx 20$** | $1/(1-k_2)$ |
| **Guadagno Totale** | Amplificazione complessiva | **$\approx 400$** | $M_1 \cdot M_2$ |
| **$k_s$ (Totale)** | Fattore di moltiplicazione della sorgente | **0.9975** | Determina la potenza prodotta |
| **$k_{eff}$ (Sistema)** | Fattore di sicurezza effettivo | **0.95** | Determinato dallo stadio più critico |

**Conclusione:**
Grazie a questa configurazione geometrica, il sistema opera con un guadagno energetico tipico di un reattore quasi critico ($k=0.9975$), ma mantiene i margini di sicurezza di un reattore profondamente sottocritico ($k=0.95$).

### **Analisi Computazionale del Burnup per l'ADS KIPT   -  2009**

#### 1. Caratteristiche del Facility (KIPT ADS)
Il sistema analizzato è un impianto sottocritico guidato da un acceleratore di elettroni, sviluppato in collaborazione tra ANL (USA) e KIPT (Ucraina).

* **Driver:** Acceleratore di elettroni da 100 kW (energia 100-200 MeV).
* **Target:** Uranio naturale. Gli elettroni colpiscono il target producendo fotoni di frenamento (Bremsstrahlung), che a loro volta generano neutroni tramite reazioni fotonucleari $(\gamma, n)$.
* **Combustibile:** LEU (Low Enriched Uranium, $\le 20\%$), tipo WWR-M2 (ossido di uranio in matrice di alluminio), geometria esagonale.
* **Refrigerante/Moderatore:** Acqua.
* **Riflettore:** Grafite.
* **Potenza:** Totale $\sim 375$ kW (di cui $\sim 260$ kW da fissione). Flusso neutronico $\sim 10^{13} n/cm^2 s$.


#### 2. Sfida Computazionale e Soluzione Ibrida
Il problema principale affrontato è che nessun codice singolo gestiva efficientemente l'intero processo:
1.  **MCNPX:** Eccellente per il trasporto di elettroni e la generazione di neutroni fotonucleari, ma all'epoca limitato nei calcoli di burnup per sistemi sottocritici con sorgente fissa.
2.  **MCB (Monte Carlo Burnup):** Eccellente per il burnup dettagliato (traccia isotopi individuali), ma incapace di trasportare elettroni.

**La Soluzione: Metodo Ibrido MCNPX-MCB**
È stata sviluppata una procedura in due step per disaccoppiare il problema:
1.  **Generazione Sorgente (MCNPX):** Si simula l'interazione elettroni-target. Ogni neutrone prodotto viene registrato in un *source file* (energia, posizione, direzione), assumendo che il target non si consumi significativamente.
2.  **Calcolo di Burnup (MCB):** Il codice MCB legge il *source file* come sorgente neutronica esterna fissa e calcola l'evoluzione del combustibile nel nocciolo sottocritico.


#### 3. Strategia di Gestione del Combustibile (Fueling Strategy)
Il burnup riduce la reattività ($k_{eff}$) e quindi il flusso neutronico (essendo la sorgente fissa). Per mantenere le prestazioni, è necessario aggiungere combustibile fresco periodicamente.

Lo studio analizza una strategia a stadi:
* **Stadio 1:** Carico iniziale di **35** assemblaggi. Operazione per **90 giorni**.
    * Il $k_{eff}$ scende da $\sim 0.98$ a $\sim 0.965$.
* **Shutdown:** Pausa di alcuni giorni per permettere il decadimento del ${}^{135}Xe$ (che fa recuperare circa 700 pcm di reattività).
* **Ricarica:** Aggiunta di **1** assemblaggio fresco (totale **36**).
    * Il $k_{eff}$ risale a $\sim 0.98$, ripristinando il flusso.
* **Stadi Successivi:** Il ciclo si ripete per raggiungere 37 e 38 assemblaggi.


#### 4. Risultati Quantitativi (Estratti)

La tabella seguente riassume l'evoluzione dei parametri chiave durante le diverse fasi di burnup simulate.

| Stadio di Burnup | Assemblaggi | Tempo (giorni) | $k_{eff}$ (BOC - Inizio) | $k_{eff}$ (EOC - Fine) | Flusso Neutroni (Core) [$n/cm^2 s$] |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **1** | 35 | 90 | $0.98062 \pm 12$ pcm | $\sim 0.964$ | $2.54 \times 10^{13}$ |
| **2** | 36 | 90 | $0.97999 \pm 12$ pcm | $\sim 0.966$ | $2.64 \times 10^{13}$ |
| **3** | 37 | 225 | $0.98173 \pm 12$ pcm | $\sim 0.966$ | $2.87 \times 10^{13}$ |

**Conclusioni dello Studio:**
* Il metodo ibrido è valido: la discrepanza tra il trasporto completo (elettroni+neutroni) e quello con sorgente neutronica fissa è trascurabile ($\sim 1\%$).
* È possibile mantenere il sistema in sicurezza ($k_{eff} \le 0.98$) e con flusso costante aggiungendo un assemblaggio alla volta dopo ogni ciclo operativo.
* Per densità di potenza basse come queste ($\sim 70 W/cm^3$), i calcoli di burnup in *k-code mode* (autovalore) approssimano bene quelli in *source mode* (sorgente fissa), semplificando le analisi future.

### **the physics of subcritical ... 2012**

Il documento è principalmente una rassegna teorica che stabilisce le basi fisiche per i sistemi ADS, ma presenta anche casi studio specifici (simulazioni su reattori veloci e dati sperimentali dal programma MUSE) per illustrare i concetti.

#### 1. Sistemi di Analisi e Configurazioni
Il paper prende in esame diverse tipologie di sistemi per discutere la fisica dei noccioli sottocritici:
* **Modello PHENIX (Teorico):** Un reattore veloce raffreddato a sodio utilizzato come benchmark per simulare il comportamento in caso di incidenti (es. inserimento di reattività) confrontando la modalità critica con quella sottocritica.
* **Programma MUSE (Sperimentale):** Esperimenti condotti presso il reattore veloce MASURCA (Cadarache) accoppiato all'acceleratore di deuteroni GENEPI. L'obiettivo è validare la fisica della sottocriticità separando gli effetti della sorgente da quelli della moltiplicazione.
* **Design Concettuali ADS:** Riferimenti a progetti come l'Energy Amplifier (Rubbia) o i sistemi raffreddati a gas/metallo liquido.


#### 2. Valori di $K_{eff}$ e $K_s$ Riportati
Il documento non fornisce un singolo set di dati univoco, ma cita diversi valori in contesti specifici:

**A. Valori di Design e Sicurezza (Range Operativi)**
Viene discusso il livello "ottimale" di sottocriticità per bilanciare sicurezza ed efficienza:
* **$K_{eff} \approx 0.95$:** Valore di riferimento usato nelle simulazioni di sicurezza per il nocciolo tipo PHENIX.
* **$K_{eff} \approx 0.95 - 0.98$:** Range operativo tipico discusso per gestire il ciclo di vita del combustibile (burn-up) e lo spegnimento sicuro ("cold" state).

**B. Valori Sperimentali (Esperimenti MUSE-3)**
Nel contesto degli esperimenti dinamici, vengono riportati livelli di sottocriticità specifici misurati in termini di pcm (per cent mille, dove $1000 \text{ pcm} \approx 1\% \Delta k/k$):
* **Configurazione REF:** $\sim 42 \text{ pcm}$ (quasi critico, $K_{eff} \approx 0.99958$).
* **Configurazione SC1:** $\sim 472 \text{ pcm}$ ($K_{eff} \approx 0.995$).
* **Configurazione SC2:** $\sim 1214 \text{ pcm}$ ($K_{eff} \approx 0.988$).
* **Configurazione SC3:** $\sim 1595 \text{ pcm}$ ($K_{eff} \approx 0.984$).

**C. Fattore $K_s$ (Source Multiplication Factor)**
Il paper definisce teoricamente $K_s$ come il fattore di moltiplicazione effettivo dei neutroni di sorgente.
* Viene evidenziato che **$K_s \neq K_{eff}$** in sistemi molto sottocritici, poiché i neutroni di sorgente hanno un'importanza diversa ($\varphi^*$) rispetto ai neutroni di fissione.
* La relazione chiave fornita è: $\varphi^* = (\frac{1}{K_{eff}} - 1) / (\frac{1}{K_s} - 1)$.

#### 3. Metodologie Utilizzate

**A. Metodologie di Calcolo (Teoriche)**
* **Equazione del Trasporto Disomogenea:** Per i sistemi sottocritici, il flusso è calcolato risolvendo l'equazione con termine di sorgente esterno $s$ ($A\phi + P\phi + s = 0$).
* **Cinetica Puntuale Estesa:** Vengono derivate equazioni cinetiche specifiche per sistemi sottocritici che includono termini per la sorgente ($P_{source}$) e la reattività generalizzata ($\rho_{gen}$), essenziali per descrivere transitori dominati dalla sorgente.

**B. Metodologie Sperimentali (Validazione)**
Il paper descrive tecniche per misurare la sottocriticità ($1-K_{eff}$) online:
* **Pulsed Neutron Source (PNS):** Analisi del decadimento temporale della popolazione neutronica dopo un impulso ($exp(-\alpha t)$) per estrarre la reattività $\rho$.
* **Modified Source Multiplication (MSM):** Tecnica statica che utilizza i ratei di conteggio dei detector per inferire la reattività, calibrando il sistema in un punto noto (vicino alla criticità).
* **Misura di $\varphi^*$:** Valutazione sperimentale dell'importanza della sorgente confrontando sorgenti diverse (es. $^{252}Cf$ vs generatori) al centro del nocciolo.
