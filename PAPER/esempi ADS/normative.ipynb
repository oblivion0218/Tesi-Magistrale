### **Analisi Normativa e Fisica per Reattore Sottocritico a Fotofissione**

Questo documento raccoglie la sintesi dei vincoli legali, normativi e fisici per la progettazione di un reattore sottocritico in Italia e nei principali paesi nucleari.


### **1. Il Vincolo Normativo sul $k_{eff}$ (Italia e Standard IAEA)**

Non esiste un articolo di legge nel Codice Italiano (D.Lgs. 101/2020) che fissi un valore numerico esplicito (es. 0,998) sul valore di $k_{eff}$. Il limite è dettato dalla prassi tecnica dell'ISIN e dagli standard IAEA.

**Il Problema di $k_{eff} \approx 1$:**
Progettare un reattore con (per esempio) $k_{eff} = 0,9999$ è legalmente insostenibile per via dell'**incertezza statistica** ($\sigma$).
* Nessun codice (MCNP, Serpent) o dato nucleare ha incertezza zero.
* Un valore di 0,9999 $\pm$ 0,0050 implica una probabilità statisticamente rilevante che il sistema sia già critico ($k \ge 1$).
* Conseguenza: Il progetto verrebbe classificato come **Reattore Critico**, perdendo i vantaggi normativi del sottocritico.

**I Limiti di Sicurezza Accettati:**
Per ottenere la licenza di "Sistema Sottocritico" (evitando la burocrazia da centrale nucleare), i valori target devono essere:
* **Stoccaggio/Trasporto:** $k_{eff} \le 0,95$ (Margine del 5%).
* **Operativo (Reattori di Ricerca/ADS):** $k_{eff} \le 0,97 - 0,98$.
* Il margine deve includere le incertezze (regola del $3\sigma$).


### **2. Panoramica Internazionale**

Nessun paese del G8 accetterebbe un design con $k_{eff} = 0,9999$ classificandolo come sottocritico.

| Paese | Ente Regolatore | Approccio Normativo |
| :--- | :--- | :--- |
| **USA** | NRC | Calcolo USL (Upper Subcritical Limit). Richiesto margine amministrativo (tipicamente 0,05 o 0,02 con controlli attivi). |
| **Francia** | ASN | Rigido. $k_{eff} \le 0,95$ standard, max 0,98 in condizioni incidentali. |
| **UK** | ONR | Approccio ALARP. Un margine < 1% è considerato rischio non giustificato. |
| **Canada** | CNSC | Richiede margine esplicito per incertezze di calcolo e fabbricazione. |
| **Russia/Cina** | Rosatom/NNSA | Possibili eccezioni solo per "Installazioni Sperimentali Strategiche", ma non classificate come sottocritiche civili standard. |



### **3. Strategia Fisica: $k_{eff}$ vs $k_s$**

La normativa regola la sicurezza intrinseca ($k_{eff}$), non il guadagno operativo ($k_s$).

**La Distinzione:**
* **$k_{eff}$ (Regolato):** Il reattore senza sorgente. Deve essere $\le 0,97$ per legge/sicurezza.
* **$k_s$ (Performance):** Il reattore con sorgente accesa. Non ha limiti legali espliciti, ma fisici.

**Formula dell'Importanza:**
$$k_s \approx \frac{k_{eff}}{1 - (1-k_{eff}) \cdot \frac{\varphi^*_{sorgente}}{\varphi^*_{fissione}}}$$

**Strategia di Design:**
1.  Mantenere $k_{eff} \approx 0,97$ per soddisfare ISIN/IAEA.
2.  Ottimizzare la geometria affinché i neutroni di sorgente abbiano un'importanza ($\varphi^*$) massima (centro del nocciolo).
3.  Risultato: Ottenere un $k_s$ operativo alto (es. 0,99) e alto guadagno energetico ($M$), pur restando legalmente sicuri.

**Attenzione:** Se $k_s \to 1$, il guadagno $M = 1/(1-k_s)$ diverge all'infinito, rendendo il sistema termicamente instabile e non controllabile.

### **4. Altre cose Legali e Burocratiche**

Oltre al $k_{eff}$, il progetto deve superare questi scogli:

1.  **Safeguards (Euratom/IAEA):**
    * Obbligo di contabilità millimetrica del materiale fissile (Uranio/Plutonio).
    * Rischio stop se il reattore è "breeder" (produce Pu-239) per timori di proliferazione.

2.  **Dual Use (Reg. UE 2021/821):**
    * Tecnologia a duplice uso (civile/militare).
    * Divieto di esportare progetti, file CAD o software di simulazione senza autorizzazione governativa.

3.  **Decommissioning (D.Lgs 101/2020):**
    * Obbligo di presentare un piano di smantellamento finanziato *prima* della costruzione.
    * Problema dell'attivazione neutronica delle strutture (acciaio radioattivo).

4.  **Doppia Licenza:**
    * Il reattore richiede approvazione sia come impianto nucleare sia come generatore di radiazioni, duplicando gli enti di controllo (Vigili del Fuoco, ASL, ISIN).

5.  **Vincolo Sismico:**
    * In Italia, la struttura deve garantire sicurezza passiva anche durante eventi sismici catastrofici (integrità del beam-pipe nel nocciolo).
